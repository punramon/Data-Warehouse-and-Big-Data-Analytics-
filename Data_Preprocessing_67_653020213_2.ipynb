> จากข้อมูลจะเห็นว่า 	pclass	survived	name	sex	sibsp	parch	และ ticket ไม่มีค่าว่าง

```python
# Select columns with int64 and float64 dtypes
numeric_cols = titanic[['age', 'sibsp', 'fare']]

# Create subplots with one row and multiple columns
fig, axes = plt.subplots(nrows=1, ncols=len(numeric_cols.columns), figsize=(len(numeric_cols.columns) * 5, 6))

# Loop through numeric columns and create box plots
for i, col in enumerate(numeric_cols.columns):
    # Calculate statistics
    min_val = numeric_cols[col].min()
    med_val = numeric_cols[col].median()
    mean_val = numeric_cols[col].mean()
    max_val = numeric_cols[col].max()

    # Identify outliers
    Q1 = numeric_cols[col].quantile(0.25)
    Q3 = numeric_cols[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = numeric_cols[col][(numeric_cols[col] < Q1 - 1.5 * IQR) | (numeric_cols[col] > Q3 + 1.5 * IQR)]

    # Print calculated values
    print(f'{col}:')
    print(f'  Min: {min_val}')
    print(f'  Median: {med_val}')
    print(f'  Mean: {mean_val}')
    print(f'  Max: {max_val}')
    #ถ้าอยากดูค่าค่อยมาเอาcommentออก print(f'  Outliers: {outliers.values}')
    print(f'  Q1: {Q1}')
    print(f'  Q3: {Q3}')
    print(f'  IQR: {IQR}')
    print('-' * 50)

    # Create box plot
    titanic.boxplot(column=col, ax=axes[i])

    # Add title
    axes[i].set_title(f'Box Plot of {col}')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```


Boxplot ของ age:
 - ค่า median อยู่ที่ 28 ปี
 - ส่วนใหญ่ของผู้โดยสารมีอายุระหว่าง 21 ถึง 39 ปี (IQR)
 - มี outliers บางส่วนที่อายุมากกว่า 65 ปี

Boxplot ของ sibsp:
 - ค่า median อยู่ที่ 0
  - หมายความว่าผู้โดยสารส่วนใหญ่ไม่ได้เดินทางมากับพี่น้องหรือคู่สมรส
 - ส่วนใหญ่ของผู้โดยสารมีจำนวนพี่น้องหรือคู่สมรสระหว่าง 0 ถึง 1 คน (IQR)
 - มี outliers บางส่วนที่มีจำนวนพี่น้องหรือคู่สมรสมากกว่า 5 คน

Boxplot ของ fare:
 - ค่า median อยู่ที่ประมาณ 14

```python
titanic.drop(columns=['body'], inplace=True)
```

* คอลัมน์ boat: คอลัมน์นี้มีค่าว่างเป็นจำนวนมาก จึงเลือกที่จะลบคอลัมน์นี้ออก -- พิจารณาจากคอลัมน์อื่นแทน

* แปลงข้อมูลเชิงคุณภาพให้เป็นตัวเลข \
    i. (10 points) ใช้ one-hot encoding ในการแปลงตัวแปร Embarked \
    ii. (10 points) แปลงตัวแปร Sex โดยให้เพศชายเป็น 0 และเพศหญิงเป็น 1 \
    iii. (10 points) สร้างตัวแปร FamilySize ซึ่งเป็นผลรวมของตัวแปร SibSp (number of siblings/spouses aboard) และ Parch (number of parents/children aboard) \
    iv. (10 points) สร้างตัวแปร IsAlone ซึ่งมีค่าเป็น 1 ถ้า FamilySize เป็น 0, และมีค่าเป็นศูนย์เมื่อเป็นอื่น

> ใช้ one-hot encoding ในการแปลงตัวแปร Embarked:
* คำสั่งนี้จะสร้างคอลัมน์ใหม่ เช่น Embarked_Q และ Embarked_S โดยการใช้ drop_first=True จะช่วยลดปัญหา multicollinearity เนื่องจากคอลัมน์หนึ่งจะถูกลบออกไป

```python
# ใช้ one-hot encoding กับตัวแปร Embarked
titanic = pd.get_dummies(titanic, columns=['embarked'], prefix='embarked', drop_first=True)
```

```python
titanic.head()
```

>  แปลงตัวแปร Sex โดยให้เพศชายเป็น 0 และเพศหญิงเป็น 1

```python
titanic['sex'] = titanic['sex'].map({'male': 0, 'female': 1})
```

```python
titanic.head()
```

> สร้างตัวแปร FamilySize ซึ่งเป็นผลรวมของตัวแปร SibSp และ Parch

```python
# สร้างตัวแปร FamilySize
titanic['FamilySize'] = titanic['sibsp'] + titanic['parch']

# ตรวจสอบผลลัพธ์
print(titanic[['sibsp', 'parch', 'FamilySize']].head())
```

> สร้างตัวแปร IsAlone ซึ่งมีค่าเป็น 1 ถ้า FamilySize เป็น 0, และมีค่าเป็นศูนย์เมื่อเป็นอื่น

```python
# สร้างตัวแปร IsAlone
titanic['IsAlone'] = (titanic['FamilySize'] == 0).astype(int)

# ตรวจสอบผลลัพธ์
print(titanic[['FamilySize', 'IsAlone']].head())
```

* ทำการ Normalize ข้อมูล \
    i. (10 points) ตัวแปร Fare และ Age ใช้ Min-Max Scaling \
    ii. (10 points) สร้างตัวแปรใหม่ Fare_Per_Person ด้วยการนำตัวแปร Fare มาหารด้วย FamilySize + 1 \
    iii. (10 points) สร้างตัวแปร AgeGroup ที่ระบุกลุ่มอายุของตัวอย่าง โดยแบ่งเป็นกลุ่มดังนี้ [0, 12), [12, 18), [18, 50) และ [50, 100)

> ใช้ Min-Max Scaling กับตัวแปร Fare และ Age
Min-Max Scaling จะปรับค่าในตัวแปรให้มีค่าอยู่ระหว่าง 0 และ 1

```python
from sklearn.preprocessing import MinMaxScaler

# สร้างตัวแปรสำหรับทำ Min-Max Scaling
scaler = MinMaxScaler()

# ทำการ Normalize ตัวแปร Fare และ Age
titanic[['fare', 'age']] = scaler.fit_transform(titanic[['fare', 'age']])

# ตรวจสอบผลลัพธ์
print(titanic[['fare', 'age']].head())
```

> สร้างตัวแปรใหม่ Fare_Per_Person ด้วยการนำตัวแปร Fare มาหารด้วย FamilySize + 1

```python
# สร้างตัวแปร Fare_Per_Person
titanic['Fare_Per_Person'] = titanic['fare'] / (titanic['FamilySize'] + 1)

# ตรวจสอบผลลัพธ์
print(titanic[['fare', 'FamilySize', 'Fare_Per_Person']].head())
```

> สร้างตัวแปร AgeGroup ที่ระบุกลุ่มอายุของตัวอย่าง โดยแบ่งเป็นกลุ่มดังนี้ [0, 12), [12, 18), [18, 50) และ [50, 100)
---> ใช้ฟังก์ชัน pd.cut() ในการแบ่งกลุ่มอายุ

```python
# สร้างตัวแปร AgeGroup
bins = [0, 12, 18, 50, 100]
labels = ['Child', 'Teenager', 'Adult', 'Senior']
titanic['AgeGroup'] = pd.cut(titanic['age'] * 100, bins=bins, labels=labels, right=False) # Scale Age back before binning

# ตรวจสอบผลลัพธ์
print(titanic[['age', 'AgeGroup']].head())
```

fare: ข้อมูลค่าโดยสารมีลักษณะเป็นโค้งเบ้ไปทางขวา (right-skewed) ซึ่งหมายความว่ามีผู้โดยสารจำนวนมากที่จ่ายค่าโดยสารในระดับต่ำ และมีจำนวนน้อยที่จ่ายค่าโดยสารสูงๆ นี่อาจบ่งบอกถึงว่ามีผู้โดยสารในชั้นที่ต่ำกว่า (เช่น ชั้น 3) มากกว่าผู้โดยสารในชั้นที่สูงกว่า

pclass: จำนวนผู้โดยสารที่อยู่ในชั้น 3 มีมากที่สุด ตามมาด้วยชั้น 1 และชั้น 2 นี่อาจบ่งบอกถึงว่ามีผู้โดยสารที่อยู่ในระดับชั้นต่ำสุด (ชั้น 3) มากกว่าระดับชั้นที่สูงกว่า (ชั้น 1 และ 2)

```python
# ใช้ crosstab เพื่อหาความสัมพันธ์ระหว่าง Survived และ Sex
survived_sex = pd.crosstab(titanic['sex'], titanic['survived'], normalize='index')
survived_sex
```

```python
sns.barplot(x='sex', y='survived', data=titanic)
plt.title('Survival Rate by Gender')
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.show()
```

> ความสัมพันธ์ระหว่าง Survived และ Age
การวิเคราะห์สถิติ:
คำนวณค่าเฉลี่ยอายุของผู้ที่รอดชีวิตและไม่รอดชีวิต

```python
# คำนวณค่าเฉลี่ยอายุของผู้ที่รอดชีวิตและไม่รอดชีวิต
age_survived = titanic.groupby('survived')['age'].mean()
age_survived
```

> การนำเสนอด้วยกราฟ:

สรุปผลการวิเคราะห์:

Survived และ Pclass: ชั้นที่นั่งมีผลต่ออัตราการรอดชีวิต ชั้นที่นั่งสูง (Pclass 1) มีอัตราการรอดชีวิตสูงกว่า

Survived และ Sex: เพศหญิงมีอัตราการรอดชีวิตสูงกว่าเพศชายอย่างชัดเจน

Survived และ Age: อายุมีความสัมพันธ์กับการรอดชีวิต โดยผู้ที่อายุน้อยและสูงอาจมีอัตราการรอดชีวิตที่แตกต่างกัน

Family Size และ Fare: ขนาดของครอบครัวมีความสัมพันธ์กับค่าโดยสาร ผู้ที่เดินทางกับครอบครัวใหญ่อาจจ่ายค่าโดยสารต่ำกว่า

> สร้าง Pivot Table เพื่อสรุปข้อมูล

```python
# สร้าง Pivot Table สรุปข้อมูล
pivot_table = titanic.pivot_table(
    index=['pclass', 'sex', 'AgeGroup'],
    values=['survived', 'fare', 'FamilySize', 'Fare_Per_Person'],
    aggfunc={
        'survived': ['size', 'sum', 'mean'],
        'fare': 'mean',
        'FamilySize': 'mean',
        'Fare_Per_Person': 'mean'
    }
)

# เปลี่ยนชื่อคอลัมน์ใน Pivot Table เพื่อให้ง่ายต่อการอ่าน
pivot_table.columns = ['Total_Passengers', 'Survived_Count', 'Survival_Rate', 'Avg_Fare', 'Avg_FamilySize', 'Avg_Fare_Per_Person']

# เพิ่มคอลัมน์การเสียชีวิต (Death Count)
pivot_table['Death_Count'] = pivot_table['Total_Passengers'] - pivot_table['Survived_Count']

# ตรวจสอบ Pivot Table ที่ได้
print(pivot_table)
```

**สรุปผลจากตารางสรุปข้อมูล**

การรอดชีวิต (Survival Rate):

อัตราการรอดชีวิตของเด็ก (Child) ในชั้นที่ 1 ค่อนข้างสูงเมื่อเทียบกับชั้นอื่น ๆ โดยเฉพาะเด็กผู้หญิงที่มีอัตราการรอดชีวิตสูงสุดที่ประมาณ 29.58%
ผู้ใหญ่ (Adult) และผู้สูงอายุ (Senior) ในชั้นที่ 1 มีอัตราการรอดชีวิตสูงกว่าผู้โดยสารในชั้นที่ 2 และ 3 อย่างชัดเจน
ผู้โดยสารชายในทุกชั้นมีอัตราการรอดชีวิตต่ำกว่าผู้โดยสารหญิง

จำนวนผู้เสียชีวิต (Death Count):

จำนวนผู้เสียชีวิตในกลุ่มผู้โดยสารชั้นที่ 3 สูงมากโดยเฉพาะในกลุ่มเด็กและวัยรุ่น โดยเฉพาะผู้ชาย
ในชั้นที่ 2 และ 3 ผู้ชายมีจำนวนผู้เสียชีวิตมากกว่าผู้หญิง
ในกลุ่มผู้โดยสารสูงอายุ (Senior) โดยเฉพาะในชั้นที่ 3 มีผู้เสียชีวิตเกือบทั้งหมด

ค่าโดยสารเฉลี่ย (Avg Fare):

ค่าโดยสารเฉลี่ยสูงสุดอยู่ในกลุ่มผู้โดยสารชั้นที่ 1 ซึ่งสะท้อนถึงความแตกต่างทางสถานะทางเศรษฐกิจและสังคม
ผู้โดยสารที่จ่ายค่าโดยสารสูงในชั้นที่ 1 มีโอกาสรอดชีวิตมากกว่าผู้โดยสารที่จ่ายค่าโดยสารน้อยกว่าในชั้นที่ 2 และ 3

ขนาดครอบครัวเฉลี่ย (Avg Family Size):

ผู้โดยสารชั้นที่ 1 มีขนาดครอบครัวเฉลี่ยน้อยกว่าเมื่อเทียบกับผู้โดยสารชั้นที่ 3 ซึ่งอาจเป็นปัจจัยที่ทำให้ผู้โดยสารชั้นที่ 1 รอดชีวิตมากกว่า
กลุ่มเด็กและวัยรุ่นมีขนาดครอบครัวเฉลี่ยสูงกว่าผู้ใหญ่และผู้สูงอายุ ซึ่งสะท้อนถึงการเดินทางร่วมกับครอบครัว
